```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
p = np.concatenate((np.zeros(50), np.ones(50)))
phat = np.concatenate((np.arange(0, 1, 0.02), np.arange(1, 0, -0.02)))
```

```python
plt.plot(phat, label="phat")
plt.plot(p, label="p")
plt.legend()
```

## Mean Squared

```python
ms = []
for i in range(len(p)):
    loss = (p[i] - phat[i])**2
    ms.append(loss)
```

```python
plt.plot(phat, label="phat")
plt.plot(p, label="p")
plt.plot(ms, label="ms")
plt.legend()
```

## Cross Entropy

```python
ce = []
for i in range(len(p)):
    loss = -np.sum(p[i]*np.log(phat[i]))
    ce.append(loss)
```

```python
plt.plot(phat, label="phat")
plt.plot(p, label="p")
plt.plot(ce, label="ce")
plt.legend()
```

## Dice Loss

```python
def dice_loss(p, phat):
    a =  2*( np.sum(p * phat) ) + 1
    b = np.sum(p+phat) + 1
    return 1 - a / b
```

```python
dl = []
for i in range(len(p)):
    loss = dice_loss(p[i], phat[i])
    dl.append(loss)
```

```python
plt.plot(phat, label="phat")
plt.plot(p, label="p")
plt.plot(dl, label="dl")
plt.legend()
```
